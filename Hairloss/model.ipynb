```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Predict Hair Fall.csv')
print(data.head())
data.describe()
```

```python
data.info()
```

```python
data.columns, data.shape
```

```python
final_data = data.drop('Id',axis=1)
print(final_data.columns)
```

```python
columns = ['Genetics','Environmental Factors','Smoking','Weight Loss ','Poor Hair Care Habits ','Hormonal Changes']
for column in columns:
    final_data[column] = final_data[column].map({'Yes': 1, 'No': 0})
final_data['Stress'] = final_data['Stress'].map({'High': 1.0, 'Moderate' : 0.5, 'Low': 0.0})   
final_data.head(2)
```

```python
from sklearn.model_selection import train_test_split
X=final_data.drop('Hair Loss',axis=1)
Y=final_data['Hair Loss']
columns_to_encode = ['Medical Conditions', 'Medications & Treatments', 'Nutritional Deficiencies ']

for column in columns_to_encode:
    frequency_map = X[column].value_counts(normalize=True)
    X[column]=X[column].map(frequency_map)

X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.7,test_size=0.3)
```

```python
# Save the training set to a CSV file by combining X_train and Y_train
train_set = pd.concat([X_train, Y_train], axis=1)
train_set.to_csv('training_set.csv', index=False)
```

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
model.score(X_train,Y_train)
model.score(X_test,Y_test)
```

Save the trained model to a file

```python
import pickle

# Save the trained model to a file
with open('prediction-ui/hair_loss_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
